# Scatter plot
Basic syntax


```python
import matplotlib.pyplot as plt
day = [1,2,3,4,5,6]
no = [1,8,9,7,7,2]
plt.scatter(day,no)
plt.title('Scatter plt',fontsize = 15,fontweight = 'bold')
plt.xlabel('Day',fontsize = 15,fontstyle = 'italic')
plt.ylabel('No',fontsize = 15,fontstyle = 'italic')
plt.show()
```

# Color

```python
colors = ['b','g','r','k','c','b']
plt.scatter(day,no,c = colors)
plt.title('Scatter plt',fontsize = 15,fontweight = 'bold')
plt.xlabel('Day',fontsize = 15,fontstyle = 'italic')
plt.ylabel('No',fontsize = 15,fontstyle = 'italic')
plt.show()
```

# We can make each dot of Diff size as well 

```python
sizes = [200,100,200,300,400,500]
colors = ['b','g','r','k','c','b']
plt.scatter(day,no,c = colors,s = sizes)
plt.title('Scatter plt',fontsize = 15,fontweight = 'bold')
plt.xlabel('Day',fontsize = 15,fontstyle = 'italic')
plt.ylabel('No',fontsize = 15,fontstyle = 'italic')
plt.show()
```

# Transpiracy of dot, alpha is lies betweem 0 to 1

```python
sizes = [200,100,200,300,400,500]
colors = ['b','g','r','k','c','b']
plt.scatter(day,no,c = colors,s = sizes,alpha = 0.5)
plt.title('Scatter plt',fontsize = 15,fontweight = 'bold')
plt.xlabel('Day',fontsize = 15,fontstyle = 'italic')
plt.ylabel('No',fontsize = 15,fontstyle = 'italic')
plt.show()
```

# We can conver the dot into diff diff shape 
## Diff shapes are 
'o': Circle
'.': Point/Dot
'x': Cross
's': Square
'^': Upward-pointing Triangle
'v': Downward-pointing Triangle
'<': Left-pointing Triangle
'>': Right-pointing Triangle
'D': Diamond
'*': Star
'H': Hexagon
'+': Plus sign
'P': Pentagon
'|': Vertical line
'_': Horizontal line

```python
sizes = [200,100,200,300,400,500]
colors = ['b','g','r','k','c','b']
plt.scatter(day,no,c = colors,s = sizes,alpha = 0.5,marker = 'p')
plt.title('Scatter plt',fontsize = 15,fontweight = 'bold')
plt.xlabel('Day',fontsize = 15,fontstyle = 'italic')
plt.ylabel('No',fontsize = 15,fontstyle = 'italic')
plt.show()
```

# Edge color and widht

```python
sizes = [200,100,200,300,400,500]
colors = ['b','g','r','k','c','b']
plt.scatter(day,no,c = colors,s = sizes,marker = 'p',edgecolor = 'k',linewidth = 2)
plt.title('Scatter plt',fontsize = 15,fontweight = 'bold')
plt.xlabel('Day',fontsize = 15,fontstyle = 'italic')
plt.ylabel('No',fontsize = 15,fontstyle = 'italic')
plt.show()
```
