```python
import matplotlib.pyplot as plt
```

BAR GRAPH 


```python
x = ["sarfuddin", "saif", "tanweer", "taha"]
y = [65,85,95,58]
plt.bar(x,y)
plt.show()
```

FOR LABELS

```python
x = ["sarfuddin", "saif", "tanweer", "taha"]
y = [65,85,95,58]
plt.xlabel("NAME")
plt.ylabel("MARKS")
plt.title("STUDENT PERFORMANCE")
plt.bar(x,y)
plt.show()
```

FONT SIZE OF LABEL


```python
x = ["sarfuddin", "saif", "tanweer", "taha"]
y = [65,85,95,58]
plt.xlabel("NAME",fontsize = 10)
plt.ylabel("MARKS",fontsize = 10)
plt.title("STUDENT PERFORMANCE",fontsize = 10)
plt.bar(x,y)
plt.show()
```

## DIFF FONT STYLE AND ITS CODE ##
Font Style	         Code
Normal	            'normal'
Bold	            'bold'
Italic	            'italic'
Bold Italic	        'bold italic'
Oblique	            'oblique' or 'italic'
Bold Oblique	    'bold oblique' or 'bold italic'
Light	            'light'
Semibold	        'semibold'

```python
x = ["sarfuddin", "saif", "tanweer", "taha"]
y = [65, 85, 95, 58]

plt.xlabel("NAME", fontsize=10, fontweight="bold") #font weight for normal and bold rest for fontstle
plt.ylabel("MARKS", fontsize=10, fontweight="bold")
plt.title("STUDENT PERFORMANCE", fontsize=12, fontstyle="italic")
plt.bar(x, y)
plt.show()
```

```python
import numpy as np

x = ["sarfuddin", "saif", "tanweer", "taha"]
y = [65, 85, 95, 58]
z = [32,45,75,38]

width = 0.2
p= np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("NAME", fontsize=10, fontweight="bold")
plt.ylabel("MARKS", fontsize=10, fontweight="bold")
plt.title("STUDENT PERFORMANCE", fontsize=12, fontstyle="italic")

plt.bar(p,y,width,color = "r",label="marks")
plt.bar(p1,z,width,color = "y",label="attendence")

plt.legend()
plt.show()
```

```python
import numpy as np

x = ["sarfuddin", "saif", "tanweer", "taha"]
y = [65, 85, 95, 58]
z = [32,45,75,38]

width = 0.2
p= np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("NAME", fontsize=10, fontweight="bold")
plt.ylabel("MARKS and Atendence", fontsize=10, fontweight="bold")
plt.title("STUDENT PERFORMANCE", fontsize=12, fontstyle="italic")

plt.bar(p,y,width,color = "r",label="marks")
plt.bar(p1,z,width,color = "y",label="attendence")

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

```python
x = ["sarfuddin", "saif", "tanweer", "taha"]
y = [65, 85, 65, 58]
z = [85, 95, 35, 58]

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

# Set the x-axis label, y-axis label, and chart title for subplot 1
ax1.set_xlabel("NAME", fontsize=10, fontweight="bold")
ax1.set_ylabel("MARKS", fontsize=10, fontweight="bold")
ax1.set_title("STUDENT PERFORMANCE", fontsize=12, fontstyle="italic")

# Create the bar chart for performance in subplot 1
ax1.bar(x, y,color="r")

# Set the x-axis label, y-axis label, and chart title for subplot 2
ax2.set_xlabel("NAME", fontsize=10, fontweight="bold")
ax2.set_ylabel("ATTENDANCE", fontsize=10, fontweight="bold")
ax2.set_title("STUDENT ATTENDANCE", fontsize=12, fontstyle="italic")

# Create the bar chart for attendance in subplot 2
ax2.bar(x, z,color="y")

# Adjust the layout of the subplots
fig.tight_layout()

# Show the chart
plt.show()







```
