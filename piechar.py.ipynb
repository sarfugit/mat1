# Pie Plot

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['c','c++','java','python']
plt.pie(x,labels = y)
plt.show()
```

# Explode

```python
x = [10,20,30,40]
y = ['c','c++','java','python']
ex = [0.2,0.1,0.0,0.0]
plt.pie(x,labels = y,explode = ex)
plt.show()
```

# Giving % value

```python
x = [10,20,30,40]
y = ['c','c++','java','python']
plt.pie(x,labels = y,explode = ex,autopct = '%0.1f%%')
plt.show()
```

## Shadow

```python
x = [10,20,30,40]
y = ['c','c++','java','python']
plt.pie(x,labels = y,explode = ex,autopct = '%0.1f%%',shadow = True)
plt.show()
```

# Coustomizing Radius

```python
x = [10,20,30,40]
y = ['c','c++','java','python']
plt.pie(x,labels = y,explode = ex,autopct = '%0.1f%%',shadow = True,radius = 2)
plt.show()
```

# Label Distance

```python
x = [10,20,30,40]
y = ['c','c++','java','python']
plt.pie(x,labels = y,explode = ex,autopct = '%0.1f%%',shadow = True,labeldistance = 0.8)
plt.show()
```

# Rotate the Circle

```python
x = [10,20,30,40]
y = ['c','c++','java','python']
plt.pie(x,labels = y,explode = ex,autopct = '%0.1f%%',shadow = True,startangle = 180)
plt.show()
```

# Text Properties: should be in Dict

```python
x = [10,20,30,40]
y = ['c','c++','java','python']
plt.pie(x,labels = y,explode = ex,autopct = '%0.1f%%',shadow = True,textprops = {'fontsize':15,'fontname':'times new roman','color':'c'})
plt.show()
```

# Edge properties , Edgeprops should also in Dic format

```python
x = [10,20,30,40]
y = ['c','c++','java','python']
plt.pie(x,labels = y,explode = ex,autopct = '%0.1f%%',shadow = True,textprops = {'fontsize':15,'fontname':'times new roman','color':'c'}, wedgeprops ={'linewidth' : 4,'edgecolor':'k'})
plt.show()
```
